## This workbook contains notes for connecting directly to a local SQLite Database

For this section, you'll need to download portal_mammals.sqlite from https://figshare.com/articles/Portal_Project_Teaching_Database/1314459

```python
import sqlite3
```

```python
conn = sqlite3.connect("data/portal_mammals.sqlite")
c = conn.cursor()
```

```python
### reading into rows
```

```python
for row in c.execute('SELECT * FROM species LIMIT 5'):
    print(row, ':', row[0])
```

```python
# working with pandas dataframes
import pandas as pd
```

```python
df_species = pd.read_sql_query("SELECT species.taxa, surveys.hindfoot_length \
          FROM species \
          JOIN surveys \
          ON species.species_id = surveys.species_id", conn)
```

```python
c.execute("CREATE TABLE taxa_hindfoot AS \
    SELECT species.taxa, surveys.hindfoot_length \
    FROM species \
    JOIN surveys \
    ON species.species_id = surveys.species_id")
```
