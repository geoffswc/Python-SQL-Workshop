```python
import pandas as pd
import glob
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())
```

```python
# using approach from 
# https://stackoverflow.com/questions/20906474/import-multiple-csv-files-into-pandas-and-concatenate-into-one-dataframe
```

```python
all_files = glob.glob('test_data/05-26_carbonhealth_and_braidhealth.csv')
df = pd.concat((pd.read_csv(f) for f in all_files), sort=False)
```

```python
pd.set_option('display.max_columns', None)
```

```python
# get the symptoms
```

```python
df_symptoms = pysqldf("""
SELECT
    covid19_test_results, labored_respiration, rhonchi,
    wheezes, cough, cough_severity,
    fever, sob, sob_severity, diarrhea, fatigue, headache,
    loss_of_smell, loss_of_taste, runny_nose, muscle_sore,
    sore_throat
FROM
    df
""")
```

```python
df_symptoms.describe()
```

```python
df.head()
```

```python
# run a simple query
sql = """
SELECT
    *
FROM
    df
WHERE
    fatigue = 1
"""

r = pysqldf(sql)
```

```python
# a group by query
```

```python
pysqldf("""
SELECT 
    covid19_test_results, fatigue, COUNT(covid19_test_results, fatigue)
FROM
    df_symptoms
GROUP BY 
    covid19_test_results, fatigue
""")
```
