```python
import pandas as pd
```

```python
rows = [
    ('A', 'ABC', 3, 1, 1), 
    ('A', 'ABC', 1, 0, 1), 
    ('A', 'DEF', 1, 0, 1), 
    ('B', 'DEF', 2, 1, 0),
    ('B', 'DEF', 3, 1, 1), 
    ('B', 'DEF', 1, 0, 1), 
    ('C', 'GHI', 1, 0, 1), 
    ('C', 'GHI', 2, 1, 0),
    ('D', 'JKL', 3, 1, 1), 
    ('D', 'JKL', 1, 0, 1), 
    ('D', 'MNO', 1, 0, 1), 
    ('D', 'MNO', 2, 1, 0)
]
```

```python
headers = ['Col1', 'Col2', 'Col3', 'Col4', 'Col5']
```

```python
df = pd.DataFrame(rows, columns=headers)
```

```python
df
```

```python
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.drop_duplicates.html
```

```python
df_reduced = df.drop_duplicates(subset=['Col1', 'Col2'], ignore_index=True)
```

```python
df_reduced
```

```python
# using SQL
```

```python
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())
```

```python
pysqldf("""
WITH df_rn AS (
    SELECT 
        *, 
        ROW_NUMBER() OVER (PARTITION BY Col1, Col2) AS rn
    FROM
        df
)

SELECT * FROM df_rn WHERE rn = 1
""")
```
