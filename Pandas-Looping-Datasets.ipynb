```python
import pandas as pd
import glob
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())
import matplotlib.pyplot as plt
```

```python
all_files = glob.glob('/Users/nms-workshop/Downloads/covidclinicaldata-master/data/*.csv')
df = pd.concat((pd.read_csv(f) for f in all_files), sort=False)
```

Let's start by looking at how many records there are for each batch date

```python
df_by_batch_date = pysqldf("""
SELECT
    batch_date, COUNT(batch_date)
FROM
    df
GROUP BY 
    batch_date
""")
```

```python
df_by_batch_date
```

```python
# Let's get more granular - let's investigate positive and negative test ratios by feature over time
# starting with high risk occupation
```

```python
df_by_batch_date_result = pysqldf("""
SELECT
    batch_date, high_risk_exposure_occupation, COUNT(*) as result_count
FROM
    df
WHERE
    covid19_test_results = 'Positive'
AND 
    high_risk_exposure_occupation IS NOT NULL
GROUP BY 
    batch_date, high_risk_exposure_occupation
""")
```

```python
df_by_batch_date_result
```

Imagine you'd like to plot or chart this. It could be difficult right now, since it isn't in the format you'd like. How can we create a chart with the date, positive count, and negative count?

In pandas you can use pivot tables. Unfortunately, SQLite lacks a pivot function, though there is an advanced technique to do this.

```python
df_pivot_pandas = df_by_batch_date_result.pivot_table(index='batch_date', 
                                                      columns='high_risk_exposure_occupation', 
                                                      values='result_count')
```

```python
#df_pivot_pandas.columns = ['batch_date', 'HRE_False', 'HRE_True']
#df_pivot_pandas.reset_index(inplace=True)
```

```python
df_pivot_pandas.rename_axis(None, axis=1, inplace=True)
#data.rename_axis(None, axis=1)
```

```python
# okay... https://stackoverflow.com/questions/43756052/transform-pandas-pivot-table-to-regular-dataframe
df_pivot_pandas = df_pivot_pandas.rename_axis(None, axis=1)
```

```python
df_pivot_pandas
```

```python
df_pivot_sql = pysqldf("""
SELECT
    df1.batch_date, 
    df1.result_count AS HREO_False, 
    df2.result_count AS HREO_True
FROM
    df_by_batch_date_result df1
JOIN
    df_by_batch_date_result df2
ON
    df1.batch_date = df2.batch_date
WHERE
    df1.high_risk_exposure_occupation = 0
AND
    df2.high_risk_exposure_occupation = 1
""")
```

```python
df_pivot_sql.columns
```

It's hard to imagine you'd prefer the second approach for this specific use case, but it's worth learning how self-joins work, because pivot won't always solve your problem!

Now let's create a bar chart

```python
df_pivot_sql.plot(x="batch_date", y=["HREO_False", "HREO_True"], kind="bar", figsize=(20, 10))
plt.legend(prop={'size': 15})
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.title('Positive Tests For High Risk Exposure Occupations by Batch Date', fontsize=20)
#plt.gca().invert_yaxis()
plt.xlabel('')
plt.show()
```
